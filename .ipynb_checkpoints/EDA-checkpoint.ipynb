```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("../Analytics Vidhya/dataset/train_ctrUa4K.csv")
```

```python
dataset.head()
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
numerical_features
```

```python
catagorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
catagorical_features
```

```python
feature_with_NA = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]
for feature in feature_with_NA:
    print("{} have {} % missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

## Numerical Features

### Descrete fetures

```python
descrete_feature = [feature for feature in feature_with_NA if len(dataset[feature].unique())<10 and feature not in catagorical_features]
descrete_feature
```

```python
plt.figure(figsize=(10,8))
for feature in descrete_feature:
    data = dataset.copy()
    #data.groupby(feature)['Loan_Status'].plot.bar()
    sns.countplot(x=feature,data=data)
    plt.title(feature+"Plot")
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.show()
```

### Continous Features

```python
continous_feature = [feature for feature in feature_with_NA if feature not in catagorical_features+descrete_feature]
continous_feature
```

```python
plt.figure(figsize=(10,8))
for feature in continous_feature:
    data = dataset.copy()
    #data.groupby(feature)['Loan_Status'].plot.bar()
    sns.violinplot(x='Loan_Status',y=feature,data=data)
    plt.title(feature+"Plot")
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.show()
```

```python
plt.figure(figsize=(10,8))
for feature in continous_feature:
    data = dataset.copy()
    x = data[feature]
    ax = sns.kdeplot(x, shade=False, color='crimson')
    kdeline = ax.lines[0]
    mean = x.mean()
    sdev = x.std()
    left = mean - sdev
    right = mean + sdev
    xs = kdeline.get_xdata()
    ys = kdeline.get_ydata()
    
    height = np.interp(mean, xs, ys)
    ax.vlines(mean, 0, height, color='crimson', ls=':')
    ax.fill_between(xs, 0, ys, facecolor='crimson', alpha=0.2)
    ax.fill_between(xs, 0, ys, where=(left <= xs) & (xs <= right), interpolate=True, facecolor='crimson', alpha=0.2)
    plt.title(feature+" Distribution Plot")
    plt.show()
```

```python
plt.figure(figsize=(10,8))
for feature in continous_feature:
    data = dataset.copy()
    x = data[feature]
    #ax = sns.kdeplot(x, shade=False, color='crimson')
    ax = sns.distplot(data[feature])
    plt.title(feature+" Distribution Plot")
    plt.show()
```
